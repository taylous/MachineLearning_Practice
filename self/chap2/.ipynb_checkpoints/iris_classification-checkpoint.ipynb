```python
import sklearn
print(sklearn.__version__)
```

**      **

```python
# (Iris)  load
from sklearn.datasets import load_iris
```

```python
#  
from sklearn.tree import DecisionTreeClassifier
```

```python
#  , (valid)   
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
```

```python
# iris data load
iris = load_iris()

# get iris's features array
iris_data = iris.data

print('iris data type: ', type(iris_data))
print('iris data size: ', len(iris_data))

# get iris target data
iris_label = iris.target
print('iris target values: ', iris_label)
print('iris target name: ', iris.target_names)

# convert DataFrame
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df['label'] = iris.target
iris_df.head(3)
```

**     **

```python
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label,
                                                    test_size=0.2, random_state=1)
```

**   (train) **

```python
# DecisionTreeClassfier  
df_clf = DecisionTreeClassifier(random_state=11)

#  
df_clf.fit(X_train, y_train)
```

**   (Predict) **

```python
#   DecisionTreeClassifier      
pred = df_clf.predict(X_test)
```

```python
pred
```

**  **

```python
from sklearn.metrics import accuracy_score
print(' : {0: .4f}'.format(accuracy_score(y_test, pred)))
```
